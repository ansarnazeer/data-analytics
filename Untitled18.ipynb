```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns= iris.feature_names
)
df.columns = [
    'sepal_length',
    "sepal_width",
    "petal_length",
    "petal_width"

]

df['species'] = pd.Categorical.from_codes(
    iris.target,
    iris.target_names
)

print(df.head())
```

```python
skewness = df.skew(numeric_only=True)
print(skewness)
```

```python
for col in skewness.index:
  if skewness[col] > 0:
    print(col, "-> Positive Skewed")
  elif skewness[col] < 0:
    print(col, "-> Negative Skewed")
  else:
    print(col, "-> Symmetric")
```

```python
numerical_cols = [
    'sepal_length',
    'sepal_width',
    'petal_length',
    'petal_width'
]

for col in numerical_cols:

  mean_val = df[col].mean()
  median_val = df[col].median()

  print("\n", col)
  print("Mean :", mean_val)
  print("Median :", median_val)

  if mean_val > median_val:
    print("Possible Positive Skewness")

  elif mean_val < median_val:
    print("Possible Negative Skewness")

  else:
    print("Approximately Symmetric")
```

```python
cv = (df[numerical_cols].std() / df[numerical_cols].mean()) * 100
print(cv)
```

```python
print("Feature with Highest Variability:")
print(cv.idxmax())
```

```python
correlation_matrix = df[numerical_cols].corr()
print(correlation_matrix)
```

```python
corr_pairs = correlation_matrix.unstack()

corr_pairs = corr_pairs[corr_pairs != 1]

strongest_positive = corr_pairs.idxmax()
strongest_value = corr_pairs.max()

weakest_negative = corr_pairs.idxmin()
weakest_value = corr_pairs.min()

print("Strongest Postive Correlation:")
print(strongest_positive, "->", strongest_value)

print("\nWeakest Negative Correlation:")
print(weakest_negative, "->", weakest_value)
```

```python
print("Petal length and petal width have strong positive correlation.")
print("Sepal width shows weak or negative with some features.")
```

```python
species_mean = df.groupby('species')[numerical_cols].mean()
print(species_mean)
```

```python
largest_petal = species_mean['petal_length'].idxmax()

print("Species with Largest Average Petal Length:")
print(largest_petal)
```

```python
smallest_sepal = species_mean['sepal_width'].idxmin()

print("Species with Smallest Average Sepal Width:")
print(smallest_sepal)
```

```python
print("Setosa has the smallest petal dimensions.")
print("Virginica has the largest petal dimensions.")
print("Versicolor lies between Setosa and Virginica.")
print("Petal features clearly distinguish species.")
```
